```python
# Importing the GUI Library

import tkinter
import tkinter.messagebox
import pickle

root = tkinter.Tk()
root.title(" To-Do-list Project")

# Creating the Add_task

def add_task():
    task = enter.get()
    if task!="":
        listbox_task.insert(tkinter.END,task)
        enter.delete(0,tkinter.END)
    else:
        tkinter.messagebox.showwarning(title="Warning",message="You must enter the task")
        
# Creating the Delete_Task:

def Delete_task():
    try:
        task_index = listbox_task.curselection()[0]
        listbox_task.delete(task_index)
    except:
        tkinter.messagebox.showwarning(title="Warning",message="You must select the task")
        
# Creating the load task
        
def Load_task():
    try:
        task = pickle.load(open("task.dat","rb"))
        listbox_task.delete(0,tkinter.END)
        for tasks in task:
            listbox_task.insert(tkinter.END,tasks)
    except:
        tkinter.messagebox.showwarning(title="Warning",message="Can not find task data")

# Creating the Save Task :

def Save_task():
    task =listbox_task.get(0,listbox_task.size())
    pickle.dump(task,open("task.dat","wb"))

# Creating the GUI box,frame..... :
frame = tkinter.Frame(root)
frame.pack()


listbox_task = tkinter.Listbox(frame,height=5,width=50)
listbox_task.pack(side=tkinter.LEFT)

scrollbar = tkinter.Scrollbar(frame)
scrollbar.pack(side=tkinter.RIGHT,fill=tkinter.Y)

listbox_task.config(yscrollcommand=scrollbar.set)
scrollbar.config(command=listbox_task.yview)

enter=tkinter.Entry(root,width=50)
enter.pack()

# Adding buttons to the Box.......:

button_task =  tkinter.Button(root,text="Add task",width=48,command=add_task)
button_task.pack()

button_task =  tkinter.Button(root,text="Delete task",width=48,command=Delete_task)
button_task.pack()

button_task =  tkinter.Button(root,text="Load task",width=48,command=Load_task)
button_task.pack()

button_task =  tkinter.Button(root,text="Save task",width=48,command=Save_task)
button_task.pack()


root.mainloop()
```
